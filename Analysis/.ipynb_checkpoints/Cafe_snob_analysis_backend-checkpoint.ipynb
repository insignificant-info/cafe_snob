```python
import sys
import json
from pprint import pprint
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity, linear_kernel
from pandas.io import sql
import MySQLdb
%matplotlib inline
import numpy
import math
from sqlalchemy import create_engine
```

```python
data = []
with open('crawled_cafes_9_20.json') as f:
    for line in f:
        data.append(json.loads(line))
```

```python
biz_hood = []
biz_id  = []
biz_name  = []
review_date  = []
review_text  = []
review_rating  = []
reviewer_id  = []
reviewer_name  = []
url  = []
rating = []
price_rating = []
review_count = []
attributes = []
city = []

for batch in data:
    #Use if not to remove all the blank reviews which are at the start of each group of cafes
    if (not batch['review_date']) or (not batch['reviewer_id']) or (not batch['biz_hood']) or (not batch['price_rating']):
        continue

    biz_name.append(batch['biz_name'])
    biz_hood.append(batch['biz_hood'])
    biz_id.append(batch['biz_id'][0])
    review_date.append(batch['review_date'][0])
    review_text.append(' '.join(batch['review_text']))
    review_rating.append(batch['review_rating'][0])
    reviewer_id.append(batch['reviewer_id'][0])
    reviewer_name.append(batch['reviewer_name'][0])
    url.append(batch['url'])
    rating.append(batch['rating'][0])
    price_rating.append(batch['price_rating'][0])
    review_count.append(batch['review_count'][0])
    attributes.append(batch['attributes'])
    #Hardcode the city for now
    city.append('New York')
    

review_count = map(int, review_count)
url = map(lambda x: x.split('?', 1)[0], url)
rating = map(float,rating)
```

```python
cafe_data_df = pd.DataFrame({'biz_name' : biz_name,
    'biz_hood' : biz_hood,
    'city' : city,
    'biz_id' : biz_id,
    'review_date' : review_date,
    'review_text' : review_text,
    'review_rating' : review_rating,
    'reviewer_id' : reviewer_id,
    'reviewer_name' : reviewer_name,
    'url' : url,
    'rating' : rating,
    'price_rating' : price_rating,
    'review_count' : review_count,
    'attributes' : attributes})
```

## define attirbute and characteristic keys

```python
#cafe_attributes_keys are all the attributes from the actuall attribute tags on yelp
cafe_attributes_keys = ['bike_parking','has_tv','alcohol','noise_level','wi_fi','outdoor_seating', 'good_for_kids']

#cafe_characteristic_keys are all the attributes scraped directly (not as attribute tags)
cafe_characteristic_keys = ['review_count','price_rating','rating']
```

```python
alcohol = []
bike_parking = []
has_tv = []
noise_level = []
outdoor_seating = []
wi_fi = []
good_for_kids = []

for index, row in cafe_data_df.iterrows():
    attribute_dict = row.attributes
    
    if cafe_attributes_keys[0] in attribute_dict:
        bike_parking.append(attribute_dict[cafe_attributes_keys[0]])
    else:
        bike_parking.append(float('nan'))
        
    if cafe_attributes_keys[1] in attribute_dict:
        has_tv.append(attribute_dict[cafe_attributes_keys[1]])
    else:
        has_tv.append(float('nan'))
        
    if cafe_attributes_keys[2] in attribute_dict:
        alcohol.append(attribute_dict[cafe_attributes_keys[2]])
    else:
        alcohol.append(float('nan'))
    
    if cafe_attributes_keys[3] in attribute_dict:
        noise_level.append(attribute_dict[cafe_attributes_keys[3]])
    else:
        noise_level.append(float('nan'))
      
    if cafe_attributes_keys[4] in attribute_dict:
        wi_fi.append(attribute_dict[cafe_attributes_keys[4]])
    else:
        wi_fi.append(float('nan'))  
        
    if cafe_attributes_keys[5] in attribute_dict:
        outdoor_seating.append(attribute_dict[cafe_attributes_keys[5]])
    else:
        outdoor_seating.append(float('nan'))
        
    if cafe_attributes_keys[6] in attribute_dict:
        good_for_kids.append(attribute_dict[cafe_attributes_keys[6]])
    else:
        good_for_kids.append(float('nan'))    
```

```python
#Add attributes to dataframe, then convert text into normalized values

cafe_data_df['bike_parking'] = pd.Series(bike_parking, index=cafe_data_df.index)
cafe_data_df['has_tv'] = pd.Series(has_tv, index=cafe_data_df.index)
cafe_data_df['alcohol'] = pd.Series(alcohol, index=cafe_data_df.index)
cafe_data_df['noise_level'] = pd.Series(noise_level, index=cafe_data_df.index)
cafe_data_df['wi_fi'] = pd.Series(wi_fi, index=cafe_data_df.index) 
cafe_data_df['outdoor_seating'] = pd.Series(outdoor_seating, index=cafe_data_df.index) 
cafe_data_df['good_for_kids'] = pd.Series(good_for_kids, index=cafe_data_df.index)

cafe_data_df.bike_parking.fillna(0, inplace=True)
cafe_data_df.has_tv.fillna(0, inplace=True)
cafe_data_df.alcohol.fillna(0, inplace=True)
cafe_data_df.noise_level.fillna(0.33, inplace=True)
cafe_data_df.wi_fi.fillna(0, inplace=True)
cafe_data_df.outdoor_seating.fillna(0, inplace=True)
cafe_data_df.good_for_kids.fillna(0, inplace=True)

try:
    cafe_data_df.price_rating = cafe_data_df['price_rating'].map(lambda x: (len(x)-1)/3)
except:
    pass
cafe_data_df.bike_parking.replace({'No':0,'Yes':1}, inplace=True)
cafe_data_df.has_tv.replace({'No':0,'Yes':1}, inplace=True)
#alcohol! Check if there is an option for liquor???
cafe_data_df.alcohol.replace({'No': float(0),'Beer & Wine Only':0.5,'Full Bar':float(1)}, inplace=True)
cafe_data_df.noise_level.replace({'Quiet': float(0),'Average':0.33,'Loud':0.66,'Very Loud': float(1), 'Very_Loud':float(1)}, inplace=True)
cafe_data_df.outdoor_seating.replace({'No':0,'Yes':1}, inplace=True)
cafe_data_df.good_for_kids.replace({'No':0,'Yes':1}, inplace=True)
cafe_data_df.wi_fi.replace({'No':0,'Paid':.5,'Free':1}, inplace=True)
```

## Define aspects groups and their accompanying words

```python
aspect_group = {}
aspect_group['methods'] = ['chemex','aeropress','pourover','siphon','grind']
aspect_group['light_roast'] = ['sour', 'scandinavian','kenya','ethiopia','acidity','origin','wave','aficionado']
aspect_group['mega_chain'] = ['fast', 'quick','efficient','bathroom']
aspect_group['work'] = ['work','laptop','quiet','wifi','tables','space','plugs','seat']
aspect_group['cozy'] = ['cute','quaint','cozy','friendly']
aspect_group['food'] = ['food','breakfast','brunch','donut','donuts','pastry','pastries']
aspect_group['alcohol_text'] = ['beer','alcohol','bar','liquor','sake']
aspect_group['rude'] = ['snooty','rude','condescending']
aspect_group['expensive'] = ['steep','expensive','pricey']
aspect_group['espresso'] = ['cortado','crema','shot','espresso']


aspect_group_keys = ['methods','light_roast','mega_chain','work','cozy','food','alcohol_text','rude','expensive','espresso']
```

```python
# Consolidate all reviews to one document per cafe
reviews_dictionary = {}
for index, row in cafe_data_df.iterrows():
    if row['biz_id'] in reviews_dictionary:
        reviews_dictionary[row['biz_id']] += row['review_text']
    else:
        reviews_dictionary[row['biz_id']] = row['review_text']
        
reviews_dict_df = pd.DataFrame(reviews_dictionary.items(), columns=['biz_id', 'review_text'])
```

```python
vectorizer = CountVectorizer(stop_words='english',
                            ngram_range=(1, 1),
                             token_pattern='[A-Za-z]+',
#                             min_df = 10,
#                             max_df = .25
                            )
features = vectorizer.fit_transform(reviews_dict_df.review_text)
features.shape
```

```python
word_vector = vectorizer.get_feature_names()
group_idx = {} 
for key, val in aspect_group.iteritems():
    for word in val:
        if key in group_idx:
            group_idx[key].append(word_vector.index(word))
        else:
            group_idx[key] = [word_vector.index(word)]
print(group_idx)
```

```python
word_counts = features.todense()

aspect_score_matrix = dict.fromkeys(aspect_group_keys)
for key, val in aspect_group.iteritems():
    aspect_score_matrix[key] = numpy.sum(word_counts[:,group_idx[key]],axis=1)
```

```python
aspect_scores = dict.fromkeys(aspect_group_keys)
for key, val in aspect_group.iteritems():    
    score_list = []
    for i in range(0,len(word_counts)):
        score_list.append(aspect_score_matrix[key][i,0])
    aspect_scores[key] = score_list
```

```python
#concat aspect_scores_df to reviews_dict_df then merge with cafe data
#gives one big df with biz data, review data, labeled info scores, and aspect scores
aspect_scores_df = pd.DataFrame(aspect_scores, index = list(reviews_dict_df.index.values))
reviews_and_word_scores = pd.concat([reviews_dict_df,aspect_scores_df],axis=1)
cafe_data_no_dupes = cafe_data_df.drop_duplicates(cols='biz_id')
biz_with_scores = pd.merge(cafe_data_no_dupes, reviews_and_word_scores,how = 'inner', left_on = 'biz_id', right_on = 'biz_id')
```

```python
#First, create a copy of biz_with_scores so I can work with the raw data
biz_with_scores_raw = biz_with_scores.copy(deep=True)

#normalize! First, weed out outliers with less than 10 reviews so they don't screw up normalization
#Just replace their review counts with 10
biz_with_scores['mod_review_count'] = biz_with_scores['review_count']
biz_with_scores['mod_review_count'][biz_with_scores['mod_review_count'] < 10] =10
```

```python
#normalize! First divide the word count by mod number of reviews. Next divide that by the column max
biz_with_scores[aspect_group_keys] = biz_with_scores[aspect_group_keys].div(numpy.log(biz_with_scores['mod_review_count']), axis = 'index')

maxes = biz_with_scores[aspect_group_keys].max(axis=0)
biz_with_scores[aspect_group_keys] = biz_with_scores[aspect_group_keys]/maxes

#more normalizing
maxes = biz_with_scores[['rating','review_count']].max(axis=0)
biz_with_scores[['rating','review_count']] = biz_with_scores[['rating','review_count']]/maxes
```

## Weight factors

```python
# Redefine cafe_attribute_keys to only include factors that I want to include
# cafe_attributes_keys = ['bike_parking','has_tv','noise_level','wi_fi','outdoor_seating', 'good_for_kids']
cafe_attributes_keys = ['noise_level','wi_fi']
```

```python
scores_only = biz_with_scores[aspect_group_keys + cafe_characteristic_keys + cafe_attributes_keys]
scores_only[aspect_group_keys] = scores_only[aspect_group_keys] * 10
scores_only[cafe_characteristic_keys] = scores_only[cafe_characteristic_keys] * 1
scores_only[cafe_attributes_keys] = scores_only[cafe_attributes_keys] * 2
scores_only[['methods','light_roast','expensive']] = scores_only[['methods','light_roast','expensive']] * 2
scores_only.columns
```

```python
#cosine similarities test before running through the loop below
similarities = cosine_similarity(scores_only.iloc[0], scores_only).flatten()
related_cafe_indices = similarities.argsort()[:-5:-1]
related_cafe_indices
related_cafe_scores = similarities[related_cafe_indices]
```

```python
match_index_names = ['first_match','second_match','third_match','fourth_match']
match_score_names = ['first_match_score','second_match_score','third_match_score','fourth_match_score']
match_url_names = ['first_match_url','second_match_url','third_match_url','fourth_match_url']

temp_row_list = []
for index in range(0,len(scores_only)):
    similarities = cosine_similarity(scores_only.iloc[index], scores_only).flatten()
    #find the indicies of top 5 matches
    related_cafe_indices = similarities.argsort()[:-6:-1]
    #find the scores of the top 5 matches
    related_cafe_scores = similarities[related_cafe_indices]
    row={}
    for index,name in enumerate(match_index_names):
        row[name] = related_cafe_indices[index+1]
    for index,name in enumerate(match_score_names):
        row[name] = related_cafe_scores[index+1]
    for index,name in enumerate(match_url_names):
        row[name] = related_cafe_indices[index+1]
    temp_row_list.append(row.copy())
```

```python
temp_df = pd.DataFrame(temp_row_list)
temp_df.head()
identification_keys = ['biz_id','biz_name','biz_hood','url']

biz_matches = pd.concat([biz_with_scores[identification_keys + aspect_group_keys + cafe_characteristic_keys + cafe_attributes_keys],temp_df],axis=1)
```

## Clustering

```python

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA


h=0.02
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(biz_matches[aspect_group_keys])

x_min, x_max = reduced_data[:, 0].min() , reduced_data[:, 0].max() 
y_min, y_max = reduced_data[:, 1].min() , reduced_data[:, 1].max() 
kmeans = KMeans(init='k-means++', n_clusters=11, n_init=10)
kmeans.fit(reduced_data)

xx, yy = numpy.meshgrid(numpy.arange(x_min, x_max, h), numpy.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(numpy.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
          extent=(xx.min(), xx.max(), yy.min(), yy.max()),
          cmap=plt.cm.Paired,
          aspect='auto', origin='lower')


centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1])
plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=2)
plt.show()
```

```python
from sklearn.cluster import AffinityPropagation
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs


h=0.02
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(biz_matches[aspect_group_keys])

centers = reduced_data
#X, labels_true = make_blobs(n_samples=, centers=centers, cluster_std=0.5,
                          # random_state=0)

##############################################################################
# Compute Affinity Propagation
af = AffinityPropagation(preference=-5., damping=0.87).fit(reduced_data)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_

n_clusters_ = len(cluster_centers_indices)


##############################################################################
# Plot result
from itertools import cycle

plt.close('all')
plt.figure(1)
plt.clf()

colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
for k, col in zip(range(n_clusters_), colors):
   class_members = labels == k
   cluster_center = reduced_data[cluster_centers_indices[k]]
   plt.plot(reduced_data[class_members, 0], reduced_data[class_members, 1], col + '.')
   plt.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
            markeredgecolor='k', markersize=14)
   for x in reduced_data[class_members]:
       plt.plot([cluster_center[0], x[0]], [cluster_center[1], x[1]], col)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.savefig('Affinity Clustering.pdf', bbox_inches='tight')
```
